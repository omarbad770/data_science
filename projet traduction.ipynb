```python
import tkinter as tk
from tkinter import ttk, messagebox
from googletrans import Translator
from PIL import Image, ImageTk, ImageFont  # Utilisez la bibliothèque Pillow (PIL) pour charger des images
import pyperclip  # Cette bibliothèque permet de copier du texte dans le presse-papiers

# Fonction pour effectuer la traduction
def traduire():
    texte_a_traduire = texte_entree.get(1.0, tk.END)
    langue_cible = langue_mapping.get(combo_langue.get())
    traduction = translator.translate(texte_a_traduire, dest=langue_cible)
    texte_traduit = traduction.text
    resultat_texte.config(state=tk.NORMAL)  # Activer la zone de texte de résultat pour modification
    resultat_texte.delete(1.0, tk.END)  # Effacer le contenu précédent
    resultat_texte.insert(tk.END, texte_traduit)  # Insérer le texte traduit
    resultat_texte.config(state=tk.DISABLED)  # Désactiver la zone de texte de résultat pour lecture
    bouton_copier.config(state=tk.NORMAL)  # Activer le bouton "Copier" après la traduction

# Fonction pour effacer le champ de texte
def effacer():
    texte_entree.delete(1.0, tk.END)
    resultat_texte.config(state=tk.NORMAL)  # Activer la zone de texte de résultat pour modification
    resultat_texte.delete(1.0, tk.END)  # Effacer le contenu précédent
    resultat_texte.config(state=tk.DISABLED)  # Désactiver la zone de texte de résultat pour lecture
    bouton_copier.config(state=tk.DISABLED)  # Désactiver le bouton "Copier" après l'effacement

# Fonction pour copier le texte traduit dans le presse-papiers
def copier_texte():
    texte_traduit = resultat_texte.get(1.0, tk.END)
    pyperclip.copy(texte_traduit)  # Copier le texte dans le presse-papiers
    messagebox.showinfo("Copié", "Le texte traduit a été copié dans le presse-papiers.")

# Créez une instance du traducteur
translator = Translator()

# Créez une fenêtre Tkinter
fenetre = tk.Tk()
fenetre.title("Traducteur automatique")

# Charger une image avec Pillow (PIL)
image_pil = Image.open("google-traduction.png")  # Remplacez par le chemin de votre image
# Réduire la taille de l'image
largeur, hauteur = image_pil.size
nouvelle_largeur = 200  # Spécifiez la nouvelle largeur souhaitée
facteur_redimensionnement = nouvelle_largeur / largeur
nouvelle_hauteur = int(hauteur * facteur_redimensionnement)
image_pil = image_pil.resize((nouvelle_largeur, nouvelle_hauteur), Image.ANTIALIAS)

# Convertir l'image PIL en format Tkinter compatible
image_tk = ImageTk.PhotoImage(image_pil)

# Afficher l'image à l'aide d'un label
image_label = tk.Label(fenetre, image=image_tk)
image_label.grid(row=0, column=0, columnspan=2, padx=5, pady=5)  # Utilisez gridgrid

# Agrandir la police de caractères pour les labels et les boutons
police_label = ("Helvetica", 10)

# Libellé pour le texte d'entrée
label_texte = tk.Label(fenetre, text="Entrez le texte à traduire:", font=police_label)
label_texte.grid(row=1, column=0, sticky="w", padx=10, pady=5 ) # Utilisez grid

# Créez une zone de texte pour l'entrée de l'utilisateur avec une barre de défilement
texte_entree = tk.Text(fenetre, wrap=tk.WORD, width=40, height=5, font=police_label)
texte_entree.grid(row=1, column=1, padx=10, pady=5)  # Utilisez grid

# Créez une barre de défilement pour la zone de texte d'entrée
scrollbar_entree = tk.Scrollbar(fenetre, command=texte_entree.yview)
scrollbar_entree.grid(row=1, column=2, sticky='ns')  # Utilisez grid

# Configurez la barre de défilement pour fonctionner avec la zone de texte d'entrée
texte_entree.config(yscrollcommand=scrollbar_entree.set)

# Libellé pour la langue cible
label_langue = tk.Label(fenetre, text="Langue cible:", font=police_label)
label_langue.grid(row=2, column=0, sticky="w", padx=10, pady=5)  # Utilisez grid

# Menu déroulant pour sélectionner la langue cible
langue_mapping = {
    "Français": "fr",
    "Anglais": "en",
    "Espagnol": "es",
    "Allemand": "de",
    "Arabe": "ar",
}

langues_disponibles = list(langue_mapping.keys())
combo_langue = ttk.Combobox(fenetre, values=langues_disponibles, font=police_label)
combo_langue.grid(row=2, column=1, padx=10, pady=5)  # Utilisez grid
combo_langue.set("Français")

# Agrandir la police de caractères pour les boutons
police_bouton = ("Helvetica", 12)

# Bouton de traduction avec une couleur de fond personnalisée
bouton_traduire = tk.Button(fenetre, text="Traduire", command=traduire, bg="green", padx=10, pady=5, font=police_bouton)
bouton_traduire.grid(row=3, column=0, columnspan=2, padx=10, pady=5)  # Utilisez grid

# Bouton pour effacer le champ de texte avec une couleur de fond personnalisée
bouton_effacer = tk.Button(fenetre, text="Effacer", command=effacer, bg="red", padx=10, pady=5, font=police_bouton)
bouton_effacer.grid(row=4, column=0, columnspan=2, padx=10, pady=5)  # Utilisez grid

# Bouton pour copier le texte traduit
bouton_copier = tk.Button(fenetre, text="Copier", command=copier_texte, padx=10, pady=5, font=police_bouton)
bouton_copier.grid(row=5, column=0, columnspan=2, padx=10, pady=5)  # Utilisez grid
bouton_copier.config(state=tk.DISABLED)  # Désactiver le bouton "Copier" initialement

# Créez une zone de texte pour afficher le texte traduit avec une barre de défilement
resultat_texte = tk.Text(fenetre, wrap=tk.WORD, width=40, height=7, font=police_label)
resultat_texte.grid(row=0, column=3, rowspan=8, padx=15, pady=15)  # Utilisez grid

# Créez une barre de défilement pour la zone de texte de résultat
scrollbar_resultat = tk.Scrollbar(fenetre, command=resultat_texte.yview)
scrollbar_resultat.grid(row=1, column=4, rowspan=6, sticky='ns')  # Utilisez grid

# Configurez la barre de défilement pour fonctionner avec la zone de texte de résultat
resultat_texte.config(yscrollcommand=scrollbar_resultat.set)

# Désactivez la zone de texte de résultat pour éviter la modification du texte traduit par l'utilisateur
resultat_texte.config(state=tk.DISABLED)

# Lancer la boucle principale de l'interface graphique
fenetre.mainloop()
```
